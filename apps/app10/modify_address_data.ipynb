```python
import pandas
df1 = pandas.read_csv('addresses_org.csv')
df1
```

```python
df2 = df1
df2['FullAddress'] = df2['Address'] + ', ' + df2['Locality']
df2 = df2.drop('Address', 1)
df2
```

```python
df3 = df2
df3['Address'] = df3['FullAddress']
df3 = df3.drop('FullAddress', 1)
df3
```

```python
df4 = df3
df4 = df4.set_index('Address')
df4
```

```python
grouped = df4.groupby(level=0)  
df5 = grouped.last() 
df5
```

```python
df5.to_csv('addresses.csv')
```
